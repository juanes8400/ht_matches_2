```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import re
import pandas as pd
import string
```

```python
def read_ht_matches(initial_id, n_id):
  for i in range(627485327,627485327+8000):
    
    #Inicialize the table
    df_ht = pd.DataFrame(columns={'MatchId','Home midfield',
                              'Away midfield','Home Right Defence',
                              'Away Right Defence','Home Central Defence',
                              'Away Central Defence','Home Left Defence',
                              'Away Left Defence','Home Right Attack',
                              'Away Right Attack','Home Central Attack',
                              'Away Central Attack','Home Left Attack',
                              'Away Left Attack','Home Indirect Defence',
                              'Away Indirect Defence','Home Indirect Attack',
                              'Away Indirect Attack','Home Attitude','Away Attitude','Home Tactic',
                              'Away Tactic','Home Tactic Level','Away Tactic Level'})
    
    url2 = 'https://www88.hattrick.org/Club/Matches/Match.aspx?matchID='+str(i)
    response = requests.get(url2)
    time.sleep(0.01)
    soup = BeautifulSoup(response.text, 'html.parser')
    s1 = soup.findAll('td')[3]
    s2 = soup.findAll('td')[4]
    s3 = soup.findAll('td')[10]
    s4 = soup.findAll('td')[11]
    s5 = soup.findAll('td')[17]
    s6 = soup.findAll('td')[18]
    s7 = soup.findAll('td')[24]
    s8 = soup.findAll('td')[25]
    s9 = soup.findAll('td')[31]
    s10 = soup.findAll('td')[32]
    s11 = soup.findAll('td')[38]
    s12 = soup.findAll('td')[39]
    s13 = soup.findAll('td')[45]
    s14 = soup.findAll('td')[46]
    s15 = soup.findAll('td')[54]
    s16 = soup.findAll('td')[55]
    s17 = soup.findAll('td')[61]
    s18 = soup.findAll('td')[62]
    s19 = soup.findAll('td')[67]
    s20 = soup.findAll('td')[68]
    s21 = soup.findAll('td')[70]
    s22 = soup.findAll('td')[71]
    s23 = soup.findAll('td')[75]
    s24 = soup.findAll('td')[76]
    
    df_ht = df_ht.append({'MatchId': i,
                          'Home midfield': float(s1.contents[0]),
                          'Away midfield': float(s2.contents[0]),
                          'Home Right Defence': float(s3.contents[0]),
                          'Away Right Defence': float(s4.contents[0]),
                          'Home Central Defence': float(s5.contents[0]),
                          'Away Central Defence': float(s6.contents[0]),
                          'Home Left Defence': float(s7.contents[0]),
                          'Away Left Defence': float(s8.contents[0]),
                          'Home Right Attack': float(s9.contents[0]),
                          'Away Right Attack': float(s10.contents[0]),
                          'Home Central Attack': float(s11.contents[0]),
                          'Away Central Attack': float(s12.contents[0]),
                          'Home Left Attack': float(s13.contents[0]),
                          'Away Left Attack': float(s14.contents[0]),
                          'Home Indirect Defence': float(s15.contents[0]),
                          'Away Indirect Defence': float(s16.contents[0]),
                          'Home Indirect Attack': float(s17.contents[0]),
                          'Away Indirect Attack': float(s18.contents[0]),
                          'Home Attitude': (s19.contents[0]),
                          'Away Attitude': (s20.contents[0]),
                          'Home Tactic': s21.contents[0],
                          'Away Tactic': s22.contents[0],
                          'Home Tactic Level': s23.contents[0],
                          'Away Tactic Level': s24.contents[0]},
                         ignore_index=True)
    df_ht.index = df_ht.MatchId
    df_ht = df_ht.drop(columns = ['MatchId'])

    return(df_ht)
```
